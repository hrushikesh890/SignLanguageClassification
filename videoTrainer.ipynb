```
equalto = np.load('drive/opticalvids/nhequalto_opt.npy')
gthan = np.load('drive/opticalvids/nhgthan_opt.npy')
lthan = np.load('drive/opticalvids/nhlthan_opt.npy')

print(equalto.shape)
print(gthan.shape)
print(lthan.shape)
```

```
y = []
for i in range(0,61):
  y.append(0)
for i in range(0,52):
  y.append(1)
for i in range(0,55):
  y.append(2)

y = np.array(y)
y = to_categorical(y,num_classes=3)

x = np.concatenate((equalto,gthan),axis=0)
print(x.shape)
x = np.concatenate((x,lthan),axis=0)
print(x.shape,len(y))

equalto = None
gthan = None
lthan = None


from sklearn.utils import shuffle
x, y = shuffle(x,y,random_state=10)
```

```
from keras.layers import Dense, GlobalAveragePooling2D, TimeDistributed
from keras.layers.recurrent import LSTM

model = Sequential()

model.add(TimeDistributed(Conv2D(32, (3, 3),strides=(2,2), activation='relu', padding='same'), input_shape=(20,480,640,3)))
model.add(TimeDistributed(Conv2D(32, (3,3), activation='relu', padding='same')))
model.add(TimeDistributed(MaxPooling2D((2, 2), strides=(2, 2))))
#model.add(Dropout(0.3))

model.add(TimeDistributed(Conv2D(64, (3,3),padding='same', activation='relu')))
model.add(TimeDistributed(Conv2D(64, (3,3),padding='same', activation='relu')))
model.add(TimeDistributed(MaxPooling2D((2, 2), strides=(2, 2))))

#model.add(Dropout(0.3))

model.add(TimeDistributed(Conv2D(128, (3,3),padding='same', activation='relu')))
model.add(TimeDistributed(Conv2D(128, (3,3),padding='same', activation='relu')))
model.add(TimeDistributed(MaxPooling2D((2, 2), strides=(2, 2))))

model.add(TimeDistributed(Conv2D(256, (3,3),padding='same', activation='relu')))
model.add(TimeDistributed(Conv2D(256, (3,3),padding='same', activation='relu')))
model.add(TimeDistributed(MaxPooling2D((2, 2), strides=(2, 2))))

#model.add(Dropout(0.3))

model.add(TimeDistributed(Conv2D(512, (3,3),padding='same', activation='relu')))
model.add(TimeDistributed(Conv2D(512, (3,3),padding='same', activation='relu')))
model.add(TimeDistributed(MaxPooling2D((2, 2), strides=(2, 2))))

model.add(TimeDistributed(Flatten()))

model.add(Dropout(0.25))
model.add(LSTM(512))
model.add(Dropout(0.25))
model.add(Dense(3, activation='softmax'))
```

```
lrate = 0.00001
nepochs =15
dec = lrate/nepochs
rms = RMSprop(lr=lrate,decay = dec)
sgd = SGD(lr=lrate, decay=dec, momentum=0.9, nesterov=True)
adam= Adam(lr=0.0001)
model.compile(optimizer=rms, loss='categorical_crossentropy',metrics=["accuracy"])
model.summary()


# In[ ]:

#model.load_weights('drive/training_params/video(224,224_1).hdf5')
from keras import callbacks
filename='drive/training_params/model_train_video1.csv'
csv_log=callbacks.CSVLogger(filename, separator=',', append=False)
filepath="drive/training_params/video(224,224_1).hdf5"
checkpoint = callbacks.ModelCheckpoint(filepath, monitor='val_loss', verbose=1, save_best_only=True, mode='min')
callbacks_list = [csv_log,checkpoint]
#hist=model.fit_generator(datagen.flow(X_train,y_train, batch_size=4),steps_per_epoch=1000, epochs=15,verbose = 1,validation_data=datagen.flow(X_test,y_test, batch_size=4),validation_steps=612,callbacks=callbacks_list)
hist=model.fit(x,y, batch_size=1, epochs=nepochs,verbose = 1,validation_split=0.2,callbacks=callbacks_list)
```
